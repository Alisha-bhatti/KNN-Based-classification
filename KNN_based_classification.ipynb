```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt


# Step 1: Create a simple dataset
data = {
    'Weight': [120, 140, 170, 200, 80, 100, 150, 180],
    'Sweetness': [7, 8, 9, 9, 5, 6, 8, 7],
    'Fruit': ['Apple', 'Apple', 'Apple', 'Apple', 'Orange', 'Orange', 'Orange', 'Orange']
}


df = pd.DataFrame(data)
print("Dataset:\n", df)


# Step 2: Split features and target
X = df[['Weight', 'Sweetness']]
y = df['Fruit']


# Step 3: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)


# Step 4: Create KNN model (k=3)
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)


# Step 5: Predict on test data
y_pred = model.predict(X_test)
print("\nPredictions:", y_pred)


# Step 6: Evaluate accuracy
print("\nAccuracy:", accuracy_score(y_test, y_pred))


# Step 7: Predict a new fruit
new_fruit = [[160, 8]]
predicted = model.predict(new_fruit)
print("\nNew Fruit with (Weight=160, Sweetness=8) is predicted as:", predicted[0])


plt.scatter(df['Weight'], df['Sweetness'], c=(df['Fruit']=='Apple'), cmap='coolwarm', s=100)
plt.xlabel('Weight (g)')
plt.ylabel('Sweetness')
plt.title('Apple vs Orange Classification')
plt.show()
```
